 #### use machine learning to create a model that predicts which passengers survived the Titanic shipwreck.
 ### The Challenge
 The sinking of the Titanic is one of the most infamous shipwrecks in history.

On April 15, 1912, during her maiden voyage, the widely considered “unsinkable” RMS Titanic sank after colliding with an iceberg. Unfortunately, there weren’t enough lifeboats for everyone onboard, resulting in the death of 1502 out of 2224 passengers and crew.

While there was some element of luck involved in surviving, it seems some groups of people were more likely to survive than others.

In this challenge, we ask you to build a predictive model that answers the question: “what sorts of people were more likely to survive?” using passenger data (ie name, age, gender, socio-economic class, etc).

```python
import pandas as pd
df=pd.read_csv('C:/Users/Abhishek/Downloads/train (2).csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df[['Age','Survived']].corr()
```

```python
df.corr()
```

```python
for i in range(1,4):
    df[df['Pclass']==i]['Age'].fillna(df[df['Pclass']==i]['Age'].mean())
```

```python
import numpy as np
df['Age']=df['Age'].fillna(np.round(df['Age'].mean(),2))
df.isnull().sum()
```

```python
df=df.drop(['PassengerId','Name','PassengerId','Cabin'],axis=1)
```

```python
df=df.drop('Ticket',axis=1)
df
```

```python
df.info()
```

```python
sex_dummy=pd.get_dummies(df['Sex'],drop_first=True)
df=df.drop('Sex',axis=1)
df=pd.concat([df,sex_dummy],axis=1)
df
```

```python
Emb_dummy=pd.get_dummies(df['Embarked'],drop_first=True)
df=df.drop('Embarked',axis=1)
df=pd.concat([df,Emb_dummy],axis=1)
df
```

```python
X=df.drop('Survived',axis=1)
y=df['Survived']
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier().fit(X,y)
```

```python
print(rf.score(X,y))
```

### SVM

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
svc=SVC()
parameters={'gamma':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100],
           'C': [0.01,0.1,0.5,1,5,10,20,30,50,70,75,80,90,100]
           }
grid=GridSearchCV(svc,parameters,cv=5).fit(X,y)

print(grid.best_params_)
print(grid.best_score_)
```

```python
svc2=SVC(gamma=0.001,C=70).fit(X,y)
print(svc2.score(X,y))
```

```python
y_pred_svc=svc2.predict(test_values)
y_pred_svc
```

```python
output2=pd.DataFrame(data=y_pred_svc,columns=[['Survived']])
output2=pd.concat([test_data['PassengerId'],output2],axis=1)
output2.columns=['PassengerId','Survived']
```
