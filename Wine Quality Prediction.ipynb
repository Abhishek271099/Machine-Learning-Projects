According to experts, the wine is differentiated according to its smell, flavor, and color, but we are not a wine expert to say that wine is good or bad. <br>
Here’s the use of Machine Learning comes,we are using machine learning to check wine quality.
### Description of Dataset
volatile acidity :   Volatile acidity is the gaseous acids present in wine. <br>
fixed acidity :   Primary fixed acids found in wine are tartaric, succinic, citric, and malic <br>
residual sugar :   Amount of sugar left after fermentation. <br>
citric acid :    It is weak organic acid, found in citrus fruits naturally. <br>
chlorides :   Amount of salt present in wine. <br>
free sulfur dioxide :   So2 is used for prevention of wine by oxidation and microbial spoilage. 
total sulfur dioxide  <br>
pH :   In wine pH is used for checking acidity
density  <br>
sulphates :    Added sulfites preserve freshness and protect wine from oxidation, and bacteria. <br>
alcohol :   Percent of alcohol present in wine.


```python
import pandas as pd
df=pd.read_csv('G:/OS work/Files/wine.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Wine'].value_counts()
```

```python
# Converting the wine column to quality of wine (low,medium,high)

df['Wine']=df['Wine'].replace({1:'Low',
                       2:'Medium',
                       3:'High'})
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('Wine',axis=1),df['Wine'],test_size=0.3,random_state=0)
```

### Feature Selection

```python
from sklearn.feature_selection import mutual_info_classif
mutual_info=mutual_info_classif(X_train,y_train)
mutual_info
```

```python
mutual_info_series=pd.Series(mutual_info,index=X_train.columns)
mutual_info_series.sort_values(ascending=False)
```

```python
# Selecting Top 7 Features

from sklearn.feature_selection import SelectKBest

sel_7_cols=SelectKBest(mutual_info_classif,k=7)
sel_7_cols.fit(X_train.fillna(0),y_train)
X_train.columns[sel_7_cols.get_support()]
```

```python
# Selecting the above features

final_column=['Alcohol', 'Phenols', 'Flavanoids', 'Color.int', 'Hue', 'OD',
       'Proline']

X_train=X_train[final_column]
X_test=X_test[final_column]
    
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
model=RandomForestClassifier().fit(X_train,y_train)
model
```

```python
params={'max_depth'          : [1,2,3,4,5,6,7],
         'max_features'       : [1,2,3,4,5,6,7],
        'max_leaf_nodes'      : [2,3,4,5,6,7]}
```

```python
cv=RandomizedSearchCV(model,param_distributions=params,n_jobs=-1,cv=5).fit(X_train,y_train)
```

```python
cv.best_estimator_
```

```python
# Copying all the above parameters

model=RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
                       max_depth=6, max_features=1, max_leaf_nodes=7,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, n_estimators=10,
                       n_jobs=None, oob_score=False, random_state=None,
                       verbose=0, warm_start=False).fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy=accuracy_score(y_test,y_pred)
confusion=confusion_matrix(y_test,y_pred)
print('Model Accuracy is {:.2f}'.format(accuracy))
print('Confusion Matrix: \n',confusion)
```
